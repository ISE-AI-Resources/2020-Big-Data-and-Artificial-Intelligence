```python
grocery = []
for i in range(len(df)):
  row_list = []
  num_item = df.at[i,'Item(s)']
  for j in range(num_item):
    row_list.append(df.at[i,f'Item {j+1}'])

  grocery.append(row_list)
```

```python
from mlxtend import frequent_patterns
from mlxtend.frequent_patterns import fpgrowth

support_score=fpgrowth(data, min_support=0.02319,use_colnames=True)
```

```python
support_score
```

```python
item_list = []
for i in range(len(support_score)):
  s= support_score.itemsets[i]
  length=len(s)
  if (length==1):
    value1, = s
    item_list.append(value1)
  elif (length==2):
    value1, value2, = s
    item_list.append([value1,value2])
  # elif (length==3):
  #   value1, value2, value3, = s
  #   item_list.append([value1,value2,value3])

support_score['item_list']=item_list
support_score
```

```python
length_list = []
for i in range(len(support_score)):
  if len(support_score.item_list[i])==2:
    length_list.append(len(support_score.item_list[i]))
  else:
    length_list.append(1)
support_score['length_list']=length_list
support_score
```

```python
confident_list = []
for i in range(len(support_score)):
  if support_score.length_list[i]==1:
    confident_list.append(1)
  else:
    item=support_score.item_list[i][0]
    idx = support_score[support_score.length_list==1][support_score.item_list==item].index[0]
    confident = support_score.support[i]/support_score.at[idx,'support']
    confident_list.append(confident)

support_score['confident']=confident_list
support_score
```

```python
lift_list = []
for i in range(len(support_score)):
  if support_score.length_list[i]==1:
    lift_list.append(1)
  else:
    item1=support_score.item_list[i][0]
    item2=support_score.item_list[i][1]
    idx1 = support_score[support_score.length_list==1][support_score.item_list==item1].index[0]
    idx2 = support_score[support_score.length_list==1][support_score.item_list==item2].index[0]
    lift = support_score.support[i]/(support_score.at[idx1,'support']*support_score.at[idx2,'support'])
    lift_list.append(lift)

support_score['lift']=lift_list
support_score
```
